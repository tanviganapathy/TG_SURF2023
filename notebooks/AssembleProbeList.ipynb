```python
f_probes = open("230630-all-probe-sequences-tdel.txt", "w")
for i, row in all_probes.iterrows():
    identity = row["gene"] + "_" + row["refseq_id"][3:] + "_" + row["chromosome"] + "_" + str(i)
    f_probes.write(identity + '\t' + row["sequence"] + '\n')
f_probes.close()
```

```python
f_probes_details = open("230630-all-probe-sequences-with-additions-tdel.txt", "w")
for i, row in all_probes.iterrows():
    identity = row["gene"] + "_" + row["refseq_id"][3:] + "_" + row["chromosome"] + "_" + str(i)
    sequence = "TGACTTGNNNNNNNN" + row["sequence"] + "AGATCGGAAGAGCGTCGTGTA"
    f_probes_details.write(identity + '\t' + sequence + '\n')
f_probes_details.close()
```
