```python
import pandas as pd
pd.set_option('display.max_columns', None)
import gget
import json
import math

gapdh_dat = pd.read_table('230630-gapdh-probes-no-restrictions.txt')
```

```python
# gapdh_dat.head(len(gapdh_dat.index))
```

```python
mask = (gapdh_dat['sequence'].str.len() <= 30)
gapdh_dat = gapdh_dat.loc[mask]
gapdh_dat = gapdh_dat.sort_values(['gene_id', 'off_target'], ascending=[True, True])
gapdh_dat.head(n=300)

#data.query("'gene_id'=='EEF1A1'").head(n=20)
```

```python
gapdh_probes_dict = {}
valid_seqs = ['NM_002046']

for i in valid_seqs:
    gene = gapdh_dat.query("refseq=='" + i.split(".")[0] + "'")
    gapdh_probes_dict[i] = gene
    print(type("refseq=='" + i.split(".")[0] + "'"))
    print(i, len(gene.index))
```

```python
gapdh_print = gapdh_probes_dict["NM_002046"]
for i, row in gapdh_print.iterrows():
    print("> ID" + str(i))
    print(row["sequence"])
```
